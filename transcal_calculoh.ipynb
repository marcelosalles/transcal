# Gráficos para valores de "h" em função de "x"

## Funções para encontrar "h":

```python
def hforcadox(x=1, v=10, ro=1.1614, mi=184*10**(-7), k=0.0263):
    
    Re = (ro*v*x)/mi
    
    if Re <= 500000:
        h = 0.332*Re**(1/2)*.707**(1/3)*(k/x)
    
    else:
        h = 0.0296*Re**(4/5)*.707**(1/3)*(k/x)
    
    return h

def hforcadomedio(L=1, v=10, ro=1.1614, mi=184*10**(-7), k=0.0263):
    
    Re = (ro*v*L)/mi
    
    if Re <= 500000:
        hmedio = 0.664*Re**(1/2)*.707**(1/3)*(k/L)
        
    return hmedio

def hnatural(L=1, Ts=30, Ti=25, B=(1/300), vis=15.89*10**(-6), alpha=22.5*10**(-6), k=0.0263):
    
    Ra = (9.807*B*(Ts-Ti)*(L**3))/(vis*alpha)
    
    if Ra < 10**4:
        print('inaplicavel')
    
    else:
        
        if Ra < 10**9:
            h = 0.59*Ra**(1/4)*(k/L)
        
        else:
            
            if Ra < 10**13:
                h = 0.10*((9.807*B*(Ts-Ti)/(vis*alpha))**(1/3))*(k)
                
            else:
                print('inaplicavel')
                
    return h
```

## Funções que definem os gráficos:

```python
% matplotlib inline
import matplotlib.pyplot as plt

def grafhf(X=3, v1=1, v2=5, v3=10, ro=1.1614, mi=184*10**(-7), k=0.0263, deltax=0.01):
    
    #v1:
    
    i = 0.01
    eixox = []
    eixoy = []
    xpre = (500000*mi)/(ro*v1)
    ypre = hforcadox(xpre,v1,ro,mi,k)
    xpos = ((500000*mi)/(ro*v1))+deltax
    ypos = hforcadox(xpos,v1,ro,mi,k)
    
    while i <= X:
        
        y = hforcadox(i,v1,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('Para v=',v1,'m/s')
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')
    
    #v2:
    
    i = 0.01
    eixox = []
    eixoy = []
    xpre = (500000*mi)/(ro*v2)
    ypre = hforcadox(xpre,v2,ro,mi,k)
    xpos = ((500000*mi)/(ro*v2))+deltax
    ypos = hforcadox(xpos,v2,ro,mi,k)
    
    while i <= X:
        
        y = hforcadox(i,v2,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('Para v=',v2,'m/s')
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')

    #v3:
    
    i = 0.01
    eixox = []
    eixoy = []
    xpre = (500000*mi)/(ro*v3)
    ypre = hforcadox(xpre,v3,ro,mi,k)
    xpos = ((500000*mi)/(ro*v3))+deltax
    ypos = hforcadox(xpos,v3,ro,mi,k)
    
    while i <= X:
        
        y = hforcadox(i,v3,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('Para v=',v3,'m/s')
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')
    plt.suptitle('h para venitlação forçada em "x"', fontsize=16)
    plt.show()
    
    
def grafhfmedio(v1=1, v2=5, v3=10, ro=1.1614, mi=184*10**(-7), k=0.0263, deltax=0.01):
    
    #v1:
    
    i = 0.01
    eixox = []
    eixoy = []
    xlim = (500000*mi)/(ro*v1)
    
    while i <= xlim:
        
        y = hforcadomedio(i,v1,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    #v2:
    
    i = 0.01
    eixox = []
    eixoy = []
    xlim = (500000*mi)/(ro*v2)
    
    while i <= xlim:
        
        y = hforcadomedio(i,v2,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    #v3:
    
    i = 0.01
    eixox = []
    eixoy = []
    xlim = (500000*mi)/(ro*v3)
    
    while i <= xlim:
        
        y = hforcadomedio(i,v3,ro,mi,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    plt.suptitle('h médio p/ L = x (Rmax < 5*10^5)', fontsize=16)
    plt.show()
    
def grafhnat(X=3, Ts1=26, Ts2=30, Ts3=35, Ti=25, B=(1/300), vis=15.89*10**(-6), alpha=22.5*10**(-6), k=0.0263, deltax=.01):
    
    #Ts1:
    
    eixox = []
    eixoy = []
    xmin = (((10**4)*vis*alpha)/(9.807*B*(Ts1-Ti)))**(1/3)
    xmax = (((10**13)*vis*alpha)/(9.807*B*(Ts1-Ti)))**(1/3)
    xpre = (((10**9)*vis*alpha)/(9.807*B*(Ts1-Ti)))**(1/3)-deltax
    ypre = hnatural(xpre,Ts1,Ti,B,vis,alpha,k)
    xpos = (((10**9)*vis*alpha)/(9.807*B*(Ts1-Ti)))**(1/3)
    ypos = hnatural(xpos,Ts1,Ti,B,vis,alpha,k)
    i = xmin
    
    while i <= X:
        
        y = hnatural(i,Ts1,Ti,B,vis,alpha,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    print('para delta T =', Ts1-Ti,'K')
    print('xmin = ',xmin)
    print('xmax = ',xmax)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')

    #Ts2:
    
    eixox = []
    eixoy = []
    xmin = (((10**4)*vis*alpha)/(9.807*B*(Ts2-Ti)))**(1/3)
    xmax = (((10**13)*vis*alpha)/(9.807*B*(Ts2-Ti)))**(1/3)
    xpre = (((10**9)*vis*alpha)/(9.807*B*(Ts2-Ti)))**(1/3)-deltax
    ypre = hnatural(xpre,Ts2,Ti,B,vis,alpha,k)
    xpos = (((10**9)*vis*alpha)/(9.807*B*(Ts2-Ti)))**(1/3)
    ypos = hnatural(xpos,Ts2,Ti,B,vis,alpha,k)
    i = xmin
    
    while i <= X:
        
        y = hnatural(i,Ts2,Ti,B,vis,alpha,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    print('para delta T =', Ts2-Ti,'K')
    print('xmin = ',xmin)
    print('xmax = ',xmax)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')

        
    #Ts3:
    
    eixox = []
    eixoy = []
    xmin = (((10**4)*vis*alpha)/(9.807*B*(Ts3-Ti)))**(1/3)
    xmax = (((10**13)*vis*alpha)/(9.807*B*(Ts3-Ti)))**(1/3)
    xpre = (((10**9)*vis*alpha)/(9.807*B*(Ts3-Ti)))**(1/3)-deltax
    ypre = hnatural(xpre,Ts3,Ti,B,vis,alpha,k)
    xpos = (((10**9)*vis*alpha)/(9.807*B*(Ts3-Ti)))**(1/3)
    ypos = hnatural(xpos,Ts3,Ti,B,vis,alpha,k)
    i = xmin
    
    while i <= X:
        
        y = hnatural(i,Ts3,Ti,B,vis,alpha,k)
        
        eixox.append(i)
        eixoy.append(y)
        
        i += deltax
        
    plt.plot(eixox,eixoy)
    
    print('para delta T =', Ts3-Ti,'K')
    print('xmin = ',xmin)
    print('xmax = ',xmax)
    if X > xpre:
        plt.plot(xpos,ypos,'ro')
        desrel = (ypos/ypre)*100
        print('descontinuidade absoluta = ',ypos-ypre)
        print('descontinuidade relativa = ', '%.2f' % desrel,'%')

    plt.suptitle('h para venitlação natural com L=x', fontsize=16)
    plt.show()
```
